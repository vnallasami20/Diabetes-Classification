```python
# visuvalize and explore the data
class PlotData:
    
    def __init__(self, X, target, out_length):
        
        self.outlier_length = out_length
        self.target = target
        self.data = X
        self.numerical_column = X.select_dtypes(exclude='object').columns.to_list()
        self.categorical_column = X.select_dtypes('object').columns.to_list()  
        
        self.numerical_column.remove(self.target)
        print('Categorical features : ',self.categorical_column)
        print('\nNumerical features   : ',self.numerical_column)
        print('\nTarget name : ',self.target)
        print('\nTraining samples : {} | No.of features : {}'.format(self.data.shape[0],self.data.shape[1]))
        
    # find inter quantile range
    def iqr(self,x):
    
        return x.quantile(q=.75) - x.quantile(q=.25)
    
    # find outliers based on inter quantile range
    def outliers_count(self, x):
    
        upper_out = x.quantile(q=.75) + self.outlier_length * self.iqr(x)
        lower_out = x.quantile(q=.25) - self.outlier_length * self.iqr(x)
        
        count = len(x[x > upper_out]) + len(x[x < lower_out])

        return count
    
    # generate a summary report 
    def explore_data(self, df, print_unique='N'): 
        
        self.df = df
        
        obs = self.df.shape[0]
        types = self.df.dtypes
        counts = self.df.apply(lambda x: x.count())
        uniques = self.df.apply(lambda x: [x.unique()])
        nulls = self.df.apply(lambda x: x.isnull().sum())
        distincts = self.df.apply(lambda x: x.unique().shape[0])
        missing_ratio = (self.df.isnull().sum()/ obs) * 100
        skewness = self.df.skew()
        kurtosis = self.df.kurt() 
        outliers = self.df[self.numerical_column].apply(lambda x: self.outliers_count(x))               

        corr = self.df.corr()[self.target]
        self.eda = pd.concat([types, counts, distincts, nulls, missing_ratio,  skewness, kurtosis, outliers, corr], axis = 1, sort=False)
        corr_col = 'Corr with '  + self.target
        cols = ['Types', 'Count', 'DistinctValues', 'Nulls', 'MissingRatio',  'Skewness', 'Kurtosis', 'Outliers', corr_col ]
        self.eda.columns = cols
        
        #print(self.eda.sort_values(by='Types'),end = '')
        
        if print_unique == 'Y':            
            for index, value in uniques.items():
                if len(value[0]) < 20:
                    print('-'*110)
                    print(index, ' : ',value[0])
                else:
                    print('-'*110)
                    print(index,' : {} unique values'.format(len(value[0])))
            print('-'*110)    
        
        return self.eda.sort_values(by='Types')
    
    # find detailed stats report
    @staticmethod
    def find_stats(X, var_list):
        data= {}
        for col in var_list:   
            stat = []
            stat.append(X[col].count())
            stat.append(X[col].quantile(0.05))
            stat.append(X[col].quantile(0.25))
            stat.append(X[col].quantile(0.50))
            stat.append(X[col].quantile(0.75))
            stat.append(X[col].quantile(0.95))
            stat.append(X[col].mean())
            stat.append(X[col].median())
            stat.append(X[col].min())
            stat.append(X[col].max())
            data[col] = stat                    
        stat_df = pd.DataFrame.from_dict(data, orient='index',columns = ['Count','.05','.25','.50','.75','.95','Mean','Median','Min','Max'])
        return stat_df
    
    @staticmethod
    def diagnostic_plots(X, col, target):
        # function takes a dataframe (df) and
        # the variable of interest as arguments
        
        sns.set(rc={'figure.figsize':(18,5)})  
        
        sns.barplot(X[col], X[target],ci=None) 
        #plt.xticks(rotation=45)
        plt.axhline(y = 0.5,linestyle='--', color='red')
        plt.show()

        # histogram
        plt.subplot(1, 2, 1)
        sns.distplot(X[col], bins=50,color='blue') # ,kde_kws={'bw':0.1}
        plt.title('Histogram')

        plt.subplot(1, 2, 2)
        sns.boxplot(X[target], X[col]);
        plt.title('Boxplot')

        plt.show()

    @staticmethod
    def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
        plt.figure()
        plt.title(title)
        if ylim is not None:
            plt.ylim(*ylim)
        plt.xlabel("Training examples")
        plt.ylabel("Score")
        train_sizes, train_scores, test_scores = learning_curve(
            estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
        train_scores_mean = np.mean(train_scores, axis=1)
        train_scores_std = np.std(train_scores, axis=1)
        test_scores_mean = np.mean(test_scores, axis=1)
        test_scores_std = np.std(test_scores, axis=1)
        plt.grid()

        plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
        plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1, color="g")
        plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
        plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Validation score")

        plt.legend(loc="best")
        return plt

    @staticmethod
    def plot_validation_curve(estimator, title, X, y, param_name, param_range, ylim=None, cv=None,
                            n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
        train_scores, test_scores = validation_curve(estimator, X, y, param_name, param_range, cv)
        train_mean = np.mean(train_scores, axis=1)
        train_std = np.std(train_scores, axis=1)
        test_mean = np.mean(test_scores, axis=1)
        test_std = np.std(test_scores, axis=1)
        plt.plot(param_range, train_mean, color='r', marker='o', markersize=5, label='Training score')
        plt.fill_between(param_range, train_mean + train_std, train_mean - train_std, alpha=0.15, color='r')
        plt.plot(param_range, test_mean, color='g', linestyle='--', marker='s', markersize=5, label='Validation score')
        plt.fill_between(param_range, test_mean + test_std, test_mean - test_std, alpha=0.15, color='g')
        plt.grid() 
        plt.xscale('log')
        plt.legend(loc='best') 
        plt.xlabel('Parameter') 
        plt.ylabel('Score') 
        plt.ylim(ylim)
    
    @staticmethod
    def boxplot(df):    
        data = []
        for col in df.columns:
            data.append(go.Box(name = col, y = df[col]))
        plotly.offline.iplot(data)     
        
    @staticmethod
    def pieplot(df, target):    
        print(df[target].value_counts())
        target_dict = df[target].value_counts().to_dict()
        label = list(target_dict.keys())
        value = list(target_dict.values())
        trace=go.Pie(labels=label,values=value, hole=.3)
        fig = go.Figure(data=[trace])
        fig.update_layout(
            autosize=False,
            width=400,
            height=400           
        )
        fig.show()       
```

```python
from sklearn import metrics as skmetrics

class ClassificationMetrics:
    def __init__(self):
        self.metrics = {
            "accuracy": self._accuracy,
            "f1": self._f1,
            "precision": self._precision,
            "recall": self._recall,
            "auc": self._auc,
            "logloss": self._logloss
        }
    
    def __call__(self, metric, y_true, y_pred, y_proba=None):
        if metric not in self.metrics:
            raise Exception("Metric not implemented")
        if metric == "logloss":
            if y_proba is not None:
                return self._logloss(y_true=y_true, y_pred=y_proba)
            else:
                raise Exception("y_proba cannot be None for logloss")
        else:
            return self.metrics[metric](y_true=y_true, y_pred=y_pred)

    @staticmethod
    def _auc(y_true, y_pred):
        return skmetrics.roc_auc_score(y_true=y_true, y_score=y_pred)

    @staticmethod
    def _accuracy(y_true, y_pred):
        return skmetrics.accuracy_score(y_true=y_true, y_pred=y_pred)
    
    @staticmethod
    def _f1(y_true, y_pred):
        return skmetrics.f1_score(y_true=y_true, y_pred=y_pred)
    
    @staticmethod
    def _recall(y_true, y_pred):
        return skmetrics.recall_score(y_true=y_true, y_pred=y_pred)
    
    @staticmethod
    def _precision(y_true, y_pred):
        return skmetrics.precision_score(y_true=y_true, y_pred=y_pred)
    
    @staticmethod
    def _logloss(y_true, y_pred):
        return skmetrics.log_loss(y_true=y_true, y_pred=y_pred)
```

```python
class TrainTestSplit:
    
    def __init__(self, target, kfold=5):
        self.kfold = kfold
        self.target = target
        
    def split_train(self, df, random_state=0):
        X = df.copy()
        y = X[self.target]
        X = X.drop(labels=self.target, axis=1)        
        X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=random_state)
        return X_train,X_test,y_train,y_test
    
    def create_folds(self, df):    
        df = df.sample(frac=1,random_state = 0).reset_index(drop=True)    
        kf = StratifiedKFold(n_splits = self.kfold, shuffle=False ,random_state = 42)   
        print('Creating Stratified KFold \n')
        for fold, (train_idx, val_idx) in enumerate(kf.split(X=df, y=df[self.target])):
            print('Fold{} : {}'.format(fold,len(val_idx)))
            df.loc[val_idx, 'kfold'] = fold
        return df
    
    def split_fold(self, df, test_fold):    
        train = df.loc[df.kfold != test_fold]
        test  = df.loc[df.kfold == test_fold]
        train.drop(labels='kfold', axis=1, inplace=True)
        test.drop(labels='kfold', axis=1, inplace=True)
        print('\n Splitting training data as train and test \n')
        print('Training sample :', train.shape[0])
        print('Testing sample :', test.shape[0])
        X_train = train.drop(labels=self.target, axis=1)
        y_train = train[self.target]
        X_test = test.drop(labels=self.target, axis=1)
        y_test = test[self.target]        
        return X_train, X_test, y_train, y_test
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class NumericalImputer(BaseEstimator, TransformerMixin):
    
    def __init__(self, variables, impute_type='mean'):
        self.impute_type = impute_type
        if not isinstance(variables, list):
            self.variables = [variables]
        else:
            self.variables = variables
            
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        df = X.copy()
        if self.impute_type == 'median':            
            for col in self.variables:
                #df[col + '_NA'] = np.where(df[col].isnull(),0,1)
                df[col].fillna(df[col].median(), inplace=True)                   
        else:
            for col in self.variables:
                #df[col + '_NA'] = np.where(df[col].isnull(),0,1)
                df[col].fillna(df[col].mean(), inplace=True)   
        return df
```

```python
from sklearn.pipeline import Pipeline
IMPUTE_VARIABLES_LIST = ['Glucose','BloodPressure','BMI','SkinThickness','Insulin']
OUTLIERS_VARIABLES_LIST = ['BloodPressure','BMI','SkinThickness','Glucose','Insulin','Age']
TRANSFORM_VARIABLES_LIST = ['Age', 'BMI','DiabetesPedigreeFunction','Insulin','SkinThickness','BloodPressure','Glucose']

pipe = []
pipe.append(('NumericalImputer',NumericalImputer(variables=IMPUTE_VARIABLES_LIST, impute_type='median')))
#pipe.append(('RemoveOutliers',Winsorizer(distribution='quantiles',tail='both',fold=0.05,variables=OUTLIERS_VARIABLES_LIST)))
pipe.append(('BoxCox',vt.BoxCoxTransformer(variables = TRANSFORM_VARIABLES_LIST)))
pipe.append(('StandardScaler', StandardScaler()))
pipe.append(('LogisticRegression',LogisticRegression(C=1)))

train = pd.read_csv('input/datasets_228_482_diabetes.csv')

train[['BloodPressure','BMI','SkinThickness','Glucose','Insulin']] = \
train[['BloodPressure','BMI','SkinThickness','Glucose','Insulin']].replace(0,np.nan)

split = TrainTestSplit(target = 'Outcome')
X_train, X_test, y_train, y_test = split.split_train(train, random_state=998)

print('Train size :',X_train.shape[0])
print('Train size :',X_test.shape[0])

classification_pipe = Pipeline(pipe)
classification_pipe.fit(X_train, y_train)
```
